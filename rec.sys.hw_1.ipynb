```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
links=pd.read_csv('links.csv')
```

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings=pd.read_csv('ratings.csv')
```

```python
tags=pd.read_csv('tags.csv')
```

## Формирование общего рейтинга фильмов

```python
movies_and_ratings = movies.join(ratings.set_index('movieId'), on='movieId')
```

```python
movies_and_ratings.groupby('title')[['rating']].mean().sort_values(by='rating')
```

```python
ratings_sum = movies_and_ratings.groupby('title')[['rating']].sum()
ratings_sum
```

```python
movies_and_tags = movies.join(tags.set_index('movieId'), on='movieId')
movies_and_tags.head(10)
```

```python
tags_count = movies_and_tags.groupby('title')[['tag']].count()
tags_count ['tag'] = tags_count ['tag'] + 1
tags_count
```

## Формирование рейтинга фильмов в жанре Action

```python
Action = movies_and_ratings[movies_and_ratings['genres'] == 'Action']
Comedy = movies_and_ratings[movies_and_ratings['genres'] == 'Comedy']
Action
```

```python
ratings_sum = Action.groupby('title')[['rating']].sum().sort_values(by='rating',ascending=False)
ratings_mean = Action.groupby('title')[['rating']].mean().sort_values(by='rating',ascending=False)
ratings_sum
ratings_mean
```

```python
Result_Action = ratings_sum.join(tags_count)
Result_Action
```

```python
Result_Action ['res'] = Result_Action ['rating'] * Result_Action ['tag']
```

```python
Result_Action.sort_values(by='res',ascending=False).head(20)
```
